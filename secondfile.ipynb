```python
# FOR PRINTING THE VALUES IN UPPERCASE---
fruits = ["apple","banana","cherry","date"]
for fruit in fruits:
    print(fruit.upper())

for idx, fruit in enumerate(fruits, start=1):
    print(f"{idx}.{fruit}")
```

```python
# WHILE LOOP IN PYTHON

count = 0
while count<5:
    print(f"Count is {count}")
    count += 1
```

```python
for num in range (10):
    if num == 5:
        break    #break statement
    print(num,end=" ")
print()

for num in range (10):
    if num == 5:
        continue    #break statement
    print(num,end=" ")
print()
```

```python
#NESTED LOOPS
for i in range (1,4):
    for j in range (1,4):
        print(f"{i+j:3}", end="")
    print()
```

```python
#WAP USE A LIST COMPEHENSION TO CREATE A LIST OF ALL VOWELS FOUND IN A GIVEN SENTENCES
sentence = "She Is The Bloody killer..."

vowels = [ch for ch in sentence if ch.lower() in 'aeiou']
    
print("The vowels are: ",vowels)
```

```python
# DEFINING FUNCTION CALLING

def greet(names):
    return f"Hello, {names}"
message = greet("Ujjwal")

print(message)
```

```python
# PARAMETERS: -- default, positional, keyword, *args, **kwargs

def describe_pet(name,animal = "dog"):
    print(f"{name} is a {animal}")
describe_pet("Roy")     #use default argument
describe_pet("Ronny",animal="cat")  #keyword argument

def add_all(*args):     # Varible positional args (tuple)
    return sum(args)
print(add_all(1,2,3,4))


def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name = "Ram",age = 22, city = "Luckonw")
```
