```python
# FILE HANDLINGGGGGGGGGGGG ----------

f = open("fileh.txt", "w")
f.write("Hallo")
f.close()

# WITH OPEN WE NOT NEED TO WRITE CLOSE FUNCTION---------
with open("fileh.txt", "w") as f:
    f.write("My name is ujja. \n")
    f.write("i am a python developer. \n")
    f.write("i think this is a one of the good carrer \n")
```

```python
# FOR READING FROM THE FILE USING READ()
with open("fileh.txt", "r") as f:
    content = f.read()
    print(content)
```

```python
# FOR READING ONE LINE FROM THE FILE USING READ()
with open("fileh.txt", "r") as f:
    content = f.readline()
    print(content)
```

```python
#for remove unecessary things like spaces etc..
    # FOR READING FROM THE FILE USING READ()
with open("fileh.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
# appending to a file
with open("fileh.txt", "a") as f:
    f.write("We are adding a new content.\n")
```

```python
# FILE POINTER USING ----------- SEEK() AND TELL()
# the file pointer tracks where in the file the next read/write

with open("fileh.txt", "r") as f:
    f.read(10)            #read first 10 chracter
    print(f.tell())
```
